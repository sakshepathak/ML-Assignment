```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

# Load CSV files
naive_bayes_df = pd.read_csv("Naive Bayes_results.csv")
knn_df = pd.read_csv("KNN (k=5)_results.csv")  
decision_tree_df = pd.read_csv("Decision Tree_results.csv")
logistic_regression_df = pd.read_csv("Logistic Regression_results.csv")

# Combine results into one DataFrame
df_results = pd.concat([naive_bayes_df, knn_df, decision_tree_df, logistic_regression_df], ignore_index=True)

# Print the table
print("\nModel Performance Comparison:\n")
print(tabulate(df_results, headers="keys", tablefmt="fancy_grid"))

# Set plot style
sns.set_style("whitegrid")

# Plot comparison
plt.figure(figsize=(10, 6))
df_results.set_index("Model").plot(kind="bar", colormap="Blues", figsize=(10, 6))
plt.title("Comparison of Classification Models")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.legend(title="Metrics", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
